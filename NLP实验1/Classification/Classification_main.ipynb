```python
import os
```

```python
root_dir = './Chinese_Text_Classification_Task_Dataset/'
test_dir = root_dir + 'train/C19-Computer/' + 'C19-Computer0003.txt'
test_dir
```

```python
# 获取文件夹的名称，即分类的类别
def get_all_folders(path):
    folders = []
    for name in os.listdir(path):
        if os.path.isdir(os.path.join(path, name)):
            folders.append(name)
    return folders

folder_names = get_all_folders(root_dir + 'train/')
folder_names
```

```python
from pyhanlp import *

# 加载ICTCLAS分词器
segmenter = HanLP.newSegment().enableCustomDictionary(False)
```

```python
# 打开txt文件
text = open(test_dir, 'rb').read().decode('GB2312', 'ignore')

# 去除一些奇怪的字符
text = text.replace('\r', ' ').replace('\n', ' ').replace('\u3000', '').replace('             ', '')

# 对文件中的每一行进行分词
for line in text:
    # 进行分词
    words = segmenter.seg(line.strip())
    # 输出分词结果
    print(words)
```
